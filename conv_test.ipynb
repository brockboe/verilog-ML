```python
def conv(filt, image):
    out = np.zeros((30, 30))
    for i in range(32 - 2):
        for j in range(32 - 2):
            s = 0
            s += filt[0][0] * image[i][j]
            s += filt[0][1] * image[i][j + 1]
            s += filt[0][2] * image[i][j + 2]
            s += filt[1][0] * image[i + 1][j]
            s += filt[1][1] * image[i + 1][j + 1]
            s += filt[1][2] * image[i + 1][j + 2]
            s += filt[2][0] * image[i + 2][j]
            s += filt[2][1] * image[i + 2][j + 1]
            s += filt[2][2] * image[i + 2][j + 2]
            out[i][j] = s
    return out

filt = np.ones((3, 3))
image = np.asarray(range(32*32)).reshape((32, 32))

def convPad(filt, image):
    out = np.zeros((32, 32))
    for i in range(34 - 2):
        for j in range(34 - 2):
            s = 0
            s += filt[0][0] * image[i][j]
            s += filt[0][1] * image[i][j + 1]
            s += filt[0][2] * image[i][j + 2]
            s += filt[1][0] * image[i + 1][j]
            s += filt[1][1] * image[i + 1][j + 1]
            s += filt[1][2] * image[i + 1][j + 2]
            s += filt[2][0] * image[i + 2][j]
            s += filt[2][1] * image[i + 2][j + 1]
            s += filt[2][2] * image[i + 2][j + 2]
            out[i][j] = s
    return out
```
